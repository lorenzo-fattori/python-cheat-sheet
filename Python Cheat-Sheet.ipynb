### CONTROL FLOW
Per controllare il flusso delle istruzioni si utilizzano 3 tecniche:  
- sequenziamento: eseguire il codice in ordine dall'alto al basso  
- iterazione: eseguire un'istruzione ripetutamente (loop)  
- selezione: specifica quale percorso di codice seguire  
Un algoritmo è una serie di istruzioni passo passo per completare una task, può essere rappresentato in diversi modi


I **cicli for** servono per implementare le iterazioni quando sappiamo il numero di iterazioni da eseguire,  
Iniziano con `for`, seguito da una variabile che terrà traccia del numero di iterazioni.  
`range()` crea una sequenza di numeri partendo da 0, lo usiamo per definire il numero di iterazioni.  
La dichiarazione del loop va seguita da i due punti,  
inoltre il codice che viene ripetuto va indentato: è necessario inserire 4 spazi all'inizio della riga per ogni indentazione:

```python
for i in range(3):
    print("ciclo for", i)
```

I **cicli while** servono per le iterazioni con una condizione che va raggiunta.  
Iniziano dichiarando un counter fuori dal ciclo che conta le iterazioni ed evita loop infiniti.  
Si dichiara poi il loop con `while`, e la condizione da applicare al counter,  
La dichiarazione del loop va seguita dai due punti e il codice da ripetere va indentato

```python
counter = 0
while counter < 5:
    print("ciclo while", counter)
    counter = counter + 1
```

Le **dichiarazioni if-else** si usano per implementare la selezione in python:  
Si inizia con `if` e a seguire la condizione per cui eseguire il codice,  
è possibile anche inserire più condizioni con le operazioni logiche.  
Alla fine si può concludere con `else` per il codice da eseguire se le altre condizioni non sono vere.  
Si possono verificare più condizioni con `elif` seguito da una condizione.  
Si possono annidare queste dichiarazioni tra loro.  
Le dichiarazione if-elif-else vanno seguite dai due punti e il codice da ripetere va indentato

```python
c = 0
d = 1
e = 2
if c == d and c == e:
    print("c, d ed e sono uguali")
else:
    if c > d and c > e:
        print("c è il più grande")
    elif d > c and d > e:
        print("d è il più grande")
    elif e > c and e > d:
        print("e è il più grande")
    elif c == d:
        print("c è uguale a d")
    elif c == e:
        print("c è uguale ad e")
    elif e == d:
        print("e è uguale a d")
    else:
        print("errore")
```

Esiste tuttavia un modo più leggibile per fare ciò, il **pattern matching**,  
che in pyhton si implementa con `match case`.  
In sostanza si confronta uno o più valori dopo il `match`  
con diversi casi, nei vari `case` indentati, il codice indentato nel caso che restituisce True viene eseguito

```python
scelta = "a"

match scelta:
    case "a":
        print("Hai scelto A")
    case "b":
        print("Hai scelto B")
    case _:
        print("Scelta non valida")
```

Si definisce una funzione con `def`, seguito dal nome della funzione, le parentesi e i due punti,

Una funzione può ritornare molteplici valori.

```python
def rect(l, w):
    area = l * w
    perimeter = 2 * l + 2 * w
    invest = area > 1000
    return area, perimeter, invest

x, y, buy = rect(50, 100)
print("Area:", x, "Perimeter:", y, "Invest?", buy)
```

### USING FUNCTIONS
Le funzioni possono prendere le liste come argomenti,  
si può usare un ciclo for dentro la funzione per iterare sulla lista,  
si può prendere un input dall'utente quando viene chiamata una funzione

```python
def cambia_menu(menu):
    p = 0
    for pizza in menu:
        print("Pizza", p, ":")
        pizza = input()
        menu[p] = pizza
        p += 1

menu = ["Margherita", "Capricciosa", "Marinara"]
cambia_menu(menu)
print("Nuovo menù:", menu)
```

I **set** sono collezioni non ordinate, si creano con le parentesi graffe `{}`,  
non supportano indexing e slicing e non hanno valori duplicati.  
`add()` e `remove()` servono per aggiungere o rimuovere l'argomento dal set.  
`clear()` rimuove tutti gli elementi da un set.  
`union()` crea un nuovo set con gli elementi di entrambi i set non duplicati.  
`difference()` ritorna un set con gli elementi che sono solo nel primo set e non nel set inserito come argomento.

```python
guests = {"Anna", "Mery", "Jonathan"}
guests.add("Robert")
guests.remove("Anna")
print(guests)
guests.clear()
set1 = {"Mele", "Pere"}
set2 = {"Pere", "Banane"}
set1e2 = set1.union(set2)
print(set1e2)
set1meno2 = set1.difference(set2)
print(set1meno2)
```

I **dizionari** sono collezioni usate per conservare dati in coppie `key:value`.  
Ogni item è costituito da un value che può essere di qualsiasi tipo,  
ed è identificato da un'unica key, solitamente una stringa.  
Si creano con parentesi graffe `{}` e una virgola tra ogni key:value.  
Possono avere valori duplicati ma non chiavi duplicate(si considera l'ultima).  
`get()` serve per accedere a un singolo valore del dizionario.  
`values()`, `keys()`, `items()` servono per accedere a diversi dati dai dizionari.

```python
contatto = {
    "nome": "Lorenzo Fattori",
    "età": 21,
    "cellulare": [39, 3312163621],
    "id": 56094,
    "iscritto": False,
    "iscritto": True    # viene considerata solo questa
}
print(contatto)
id_lorenzo = contatto["id"]
print("ID:", id_lorenzo)
print("NOME:", contatto.get("nome"))
cont_keys = contatto.keys()
print("CHIAVI:", cont_keys)
print("VALORI:", contatto.values())
lorenzo = contatto.items()
print("LORENZO:", lorenzo)
```

I dizionari sono mutabili, se l'elemento che si modifica è nuovo viene aggiunto al dizionario.  
`update()` permette di cambiare o aggiungere elementi del dizionario,  
il suo argomento dev'essere un dizionario con l'elemento da modificare.  
`pop()` permette di eliminare elementi del dizionario.  
L'operatore `in` serve per sapere se un valore o una chiave è presente.  
Si può iterare su un dizionario con il ciclo for

```python
contatto["età"] = 22    # modificata
contatto["facoltà"] = "ingegneria gestionale"    # aggiunta
contatto.update({"facoltà": "digital transformation management"})
contatto.pop("cellulare")
print("Il cellulare è presente?", "cellulare" in contatto)
for i in contatto:
    print(i, ":", contatto[i])
```

### ERROR HANDLING  
I **bugs** sono errori che non per forza interrompono l'esecuzione,  
tuttavia possono causare un comportamento sbagliato del codice.  
Le **exceptions** sono errori che bloccano l'esecuzione del programma,  
possono essere:  
- `SyntaxError`;  
- `IndexError`;  
- `ValueError`;  
- `NameError`;  
- `TypeError`  

Si possono usare `try` e `except` per manipolare le exception,  
se avviene una exception nel blocco try, sarà eseguito il blocco except,  
si può specificare il tipo di exception oppure no

```python
costi = [250, 300, "240", 400]
try:
    c_tot = sum(costi)                  # codice che può causare una exception
    print("Costo totale:", c_tot)
except TypeError:
    print("Tipo di dato non valido")    # da eseguire se c'è un TypeError
except:
    print("Altro tipo di exception")    # da eseguire se c'è un'altra exception
```

`finally` è usato per eseguire del codice dopo un try/except  
indipendentemente dal fatto che sia avvenuta o no l'exception.  
`else` è usata con try/except e avviene solo se il blocco try è privo di errori.  
Si possono creare exception personalizzate con `if` e `raise`,  
è possibile aggiungere un messaggio per descrivere l'errore

```python
try:
    c_tot = sum(costi)
    print("Costo totale:", c_tot)
except TypeError:
    print("Tipo di dato non valido")
except:
    print("Altro tipo di exception")
else:
    print("Il blocco try è privo di errori")
finally:
    print("Hai bisogno di aiuto?")
```

### FUNCTIONAL PROGRAMMING  
La **programmazione funzionale** è un paradigma di programmazione che tratta le funzioni come entità principali  
e produce risultati evitando stati mutabili ed effetti collaterali.

Si può assegnare una funzione a una variabile e usarla per chiamare la funzione.  
Le funzioni che prendono altre funzioni come input e/o ritornano una funzione sono chiamate higher-order functions.  
Le funzioni pure ritornano sempre lo stesso valore per lo stesso argomento e non producono effetti collaterali

```python
def urlo(testo):
    return testo.upper() + "!"
strillare = urlo
print(strillare("Ciao"))
```

```python
def book_title(title):
    return "Book title: " + title
def info(title, func):
    return func(title)
print(info("Django Unchained", book_title))
```

Le **lambda** sono funzioni compatte e anonime usate per operazioni semplici,  
si creano in questo modo: `lambda <arg1>, <arg2>, ...: <espressione>`,  
possono essere assegnate a variabili o usate con altre funzioni

```python
area_triangolo = lambda b, h: (b*h)/2    # funzione
print("Area di un triangolo b=3 e h=4:", area_triangolo(3, 4))
area_quadrato = (lambda b, h: b*h) (2, 3)    # variabile
print("Area di un quadrato b=2 e h=3:", area_quadrato)
```

`map()` applica una funzione ad ogni elemento di un iterabile:  
`map(<funzione>, <iterabile>)`  
`filter()` filtra gli elementi da un iterabile basandosi su una certa funzione,  
entrambi accettano espressioni lambda come argomento

```python
numbers = [1, 2, 3, 4, 5]
# list crea la lista di tutti gli elementi a cui map ha applicato la funzione lambda
doubled = list(map(lambda x: x * 2, numbers))    
print(doubled)                                   
```

```python
products = ["Table", "Sofa", "Cushion", "Bookshelf", "Vase"]
filtered_prod = list(filter(lambda name: len(name) == 4, products))
print(filtered_prod)    # filtra i prodotti con lunghezza uguale a 4
```

***args** permette di passare un numero di argomenti arbitrario a una funzione:  
`def <funzione>(<argomento>, <*args>)`  

****kwargs** permette di passare argomenti keyword a una funzione (specificando il nome del parametro):  
`def <funzione>(<argomento>, <*args>, <**kwargs>)`  

*args raccoglie gli argomenti in una tupla, mentre **kwargs in un dizionario

```python
def totale(*argomenti):
    risultato = 0
    for arg in argomenti:
        risultato += arg
    return risultato
print(totale(1, 2, 3))
print(totale(2, 3, 4, 5, 6))
```

```python
def display_info(**kwargomenti):
    for key, value in kwargomenti.items():
        print(key, ":", value)
display_info(name="Alice", age=22, city="Milano")
```

I **decorators** sono usati per modificare il comportamento di una funzione senza alterare il suo codice originale.  
Si usa il simbolo `@` per applicare il decorator ad una funzione,  
lo stesso decorator può essere applicato a diverse funzioni,  
è bene usare *args e **kwargs per la creazione di un **wrapper** in un decorator,  
ovvero la funzione interna che modifica o estende il comportamento della funzione originale senza cambiarne il codice

```python
def maiuscolo(func):    # definisco il decorator
    def wrapper():    # sarà la versione modificata della funzione salve()
        mess_originale = func()
        mess_modificato = mess_originale.upper()
        return mess_modificato
    return wrapper

@maiuscolo    # uso il decorator su salve()
def salve():
    return "Salve a tutti!"
print(salve())
```

### OBJECT-ORIENTED PROGRAMMING
La **programmazione a oggetti** un modo di programmare basato su classi e oggetti.  
Le **classi** sono modelli che definiscono attributi e comportamenti (**abstraction**),  
gli **oggetti** sono istanze delle classi, sono elementi concreti e unici.  
Per definire una classe si scrive class, il nome della classe e i due punti.  
Per aggiungere attributi a una classe si definisce il metodo `__init__` seguito da `self`,  
e dagli attributi da inizializzare nell'oggetto.  
Si può accedere agli attributi di un oggetto con la dot.notation.  
I metodi sono funzioni associate a una determinata classe

```python
class Macchina:
    def __init__(self, brand, colore):
        self.brand = brand
        self.colore = colore
    def clacson(self):
        print("Beep Beep!")
mia_macchina = Macchina("Opel", "grigia")
print(mia_macchina.brand)
print(mia_macchina.clacson())
```

L'**inheritance** consente a una nuova classe di ereditare le caratteristiche di una classe già esistente,  
quando si definisce la child class è necessario includere il nome della parent class tra parentesi tonde.  
Si usa `super().__init__()` per ereditare attributi della parent class.  
Si possono aggiungere attributi e comportamenti unici a una classe figlia,  
questa è caratteristica dell'OOP è chiamata **polymorphism**.  
Infine, il **method overriding** permette di modificare un metodo ereditato

```python
class Animale:                         # parent class
    def __init__(self, nome):
        self.nome = nome
    def verso(self):
        print("Verso dell'animale")    # metodo generico

class Cane(Animale):                   # child class
    def __init__(self, nome, razza, anni):
        super().__init__(nome)
        self.razza = razza
        self.anni = anni
    def verso(self):                   # overriden method
        print("Bau Bau!")

class Gatto(Animale):                  # child class
    def __init__(self, nome, razza, anni):
        super().__init__(nome)
        self.razza = razza
        self.anni = anni
    def verso(self):
        print("Miao Miao!")

# creo le istanze
mio_cane = Cane("Pepe", "Pastore tedesco", 5)
mio_gatto = Gatto("Trilli", "Mau egiziano", 14)
animali_domestici = [mio_cane, mio_gatto]
for animale in animali_domestici:
    animale.verso()
```

Il **data hiding** è cruciale nel'OOP per un codice sicuro e organizzato.  
Usando un singolo underscore `_` indichiamo che gli attributi o metodi sono protetti, intesi per un uso interno,  
i doppi underscore `__` rendono gli attributi/metodi privati, restringendo l'accesso alla classe stessa.  
Ciò rafforza la sicurezza dei dati

```python
class Moto:
    def __init__(self, modello, contakm, targa):
        self.modello = modello
        self._contakm = contakm    # rendo contakm protetta
        self.__targa = targa       # rendo targa privata

    def leggikm(self):
        print("KM:", self._contakm)

    def leggitarga(self):
        print("TARGA:", self.__targa)

mia_moto = Moto("Ducati Monster", 23000, "FE92BG")
mia_moto.leggikm()
mia_moto.leggitarga()
print(mia_moto._contakm)
#print(mia_moto.__targa) causerebbe un errore
```

I **class methods** sono chiamati sulla classe stessa, non sulle singole istanze.  
Si definiscono con il decorator `@classmethod` e accettano l'argomento `cls`.  
Invece i `@staticmethod` sono come i class methods ma non hanno accesso alla classe

```python
class Libro:
    def __init__(self, titolo, autore):
        self.titolo = titolo
        self.autore = autore

    def descrivi_libro(self):                    # metodo regolare
        print(self.titolo, "di", self.autore)

    @classmethod
    def libri_in_serie(cls, nome_serie, num_libri):
        print("Ci sono", num_libri, "libri nella", nome_serie, "serie")

    @staticmethod
    def casa_editrice(titolo, casa):
        print("La casa editrice di", titolo, "è", casa)

# creo l'istanza di Libro
mio_libro = Libro("Harry Potter e i doni della morte", "J. K. Rowling")
mio_libro.descrivi_libro()    # metodo regolare
mio_libro.libri_in_serie("Harry Potter", 7)    # class method
Libro.casa_editrice("Harry Potter", "Mondadori")    # static method
```
